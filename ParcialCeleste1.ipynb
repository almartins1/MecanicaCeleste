```python
import pandas as pd
import numpy as np
import math

#Ejercicio 1
#Ejercicio tipo: Determinar la anomalía verdadera y el radio vector y el periodo orbital EJ1
# Datos que conocemos

a = 2.6785123       # Semieje mayor(unidades: ua)
e = 0.2543422      # Excentricidad (unidades: no tiene)
K= 0.01720209908

#Calculemos T:
#m2/m1 = 0
T=(2*np.pi*math.pow(a,3/2))/(K)



print("El periodo orbital esta dado por {0} {1}".format(T, "d"))

#Calculemos n
n=360/T

print("El movimiento medio (n) esta dado por {0} {1}".format(n, "°/d"))
print("----------------")

#------------------------------------------------------
#Calculemos t - t_0 

#Con la librería pandas ponemos T_0 como una fecha para pasarla a Juliana 
#Hora T TUC
t = pd.Timestamp(year = 2006,  month = 5, day = 5, hour = 0, minute=0, second = 0, tz = 'UTC') 
  
#Determinar r y theta para el instante 

#Con la librería pandas ponemos t como una fecha para pasarla a Juliana 
#Hora T_0 TUC
t_0 = pd.Timestamp(year = 2006,  month = 6, day = 30, hour = 0, minute=0, second = 0, tz = 'UTC') 

jult_0 = t_0.to_julian_date()
jult = t.to_julian_date()

print(t_0)
print(t)

print("----------------")

print("fecha juliana")
print(jult_0)
print(jult)
print(jult-jult_0)

print("----------------")

#Anomalía media:

M = n*(jult-jult_0)

# En caso de que M sea negativo
if(M < 0):
    M = M + 360

print("La anomalía media (M) esta dado por {0} {1}".format(M, "°"))
print("----------------")

#Calculemos la anomalía excéntrica E dada por la iteración:
#Nota np.sin de entrada son radianes por lo tanto se convierte de grados a 
#radianes para realizar la operación
E = M

#iterar mas veces si es necesario
for i in range(1,100):
    E = M + e*(180/np.pi)*(np.sin(np.deg2rad(E))) 
    if(i == 48):
        print("test de E ", E)

#Si por lo menos 8 cifras decimales coinciden entre el ultimo y el anterior
#entonces ya tenemos E

print("La anomalía excéntrica E esta dada por {0} {1}".format(E, "°"))  

print("----------------")

#Hallemos r

r = a*(1 - e*(np.cos(E*np.pi/180)))

print("El radio vector esta dado por {0} {1}".format(r, "u.a."))


print("----------------")

#Hallemos anomalia verdadera (theta)

theta = 2*np.arctan(np.sqrt((1+e)/(1-e)) * np.tan(np.deg2rad(E/2))) 

#Cambio a grados:

theta = np.rad2deg(theta)

# En caso de que theta sea negativo
if(theta < 0):
    theta = theta + 360

print("La anomalía verdadera esta dada por {0} {1}".format(theta, "°"))
```

```python
import pandas as pd
import numpy as np
import math  

#en caso de que se pase de 360 o sea negativo
def gradnorm(alpha):
  if alpha > 360:
    alpha = alpha - 360
  elif alpha < 0:
    alpha = alpha + 360
  return alpha

#Verificar el tangente inverso 
def atansi(alpha,a,b):
  if a*b < 0 and b<0:
    alpha = alpha + 180
  elif a*b < 0 and b>0:
    alpha = alpha + 360
  elif a+b < 0:
    alpha = alpha + 180
  return alpha


##Punto4

##########
#Si es asteroide m2/m1 = 0, sino mirar tabla del cuaderno y poner valor aqui
divm =  1/328900  #m2/m1
#Constante K
K = 0.01720209908


i =  np.deg2rad(0)       #Inclinación 
BigOmega = np.deg2rad(0)   #Long. nodo ascendente
SmallOmega = np.deg2rad(102.9777)
a = 1      # Semieje mayor(unidades: ua)
e = 0.016674    # Excentricidad (unidades: no tiene)
Lr = 245.26107      #Longitud  media tr

omegabar= SmallOmega+BigOmega

#hallemos longitud media

T=(2*np.pi*math.pow(a,3/2))/(K)

print("El periodo orbital esta dado por {0} {1}".format(T, "d"))

#Calculemos n
n=360/T

print("El movimiento medio (n) esta dado por {0} {1}".format(n, "°/d"))
print("----------------")


#t
#Con la librería pandas ponemos T como una fecha para pasarla a Juliana 
#Hora T TUC
t = pd.Timestamp(year = 2014,  month = 6, day = 21, hour = 0, minute=0, second = 0, tz = 'UTC')
jult = t.to_julian_date()

t_0 = pd.Timestamp(year = 2014,  month = 5, day = 28, hour = 0, minute=0, second = 0, tz = 'UTC')
jult_0 = t_0.to_julian_date()

#Longitud media usando la longitud media de referencia:

L = Lr  + n*(jult-jult_0)

print("----------------")
print("La Longitud media esta dado por {0} {1}".format(L, "°"))
print("----------------")

#Luego la anomalia media

M = L-omegabar

#Calculemos la anomalía excéntrica E dada por la iteración:
#Nota np.sin de entrada son radianes por lo tanto se convierte de grados a 
#radianes para realizar la operación
E = M

#iterar mas veces si es necesario
for l in range(1,100):
    E = M + e*(180/np.pi)*(np.sin(np.deg2rad(E))) 
    if(i == 98):
        print("test de E ", E)

#Si por lo menos 8 cifras decimales coinciden entre el ultimo y el anterior
#entonces ya tenemos E

print("La anomalía excéntrica E esta dada por {0} {1}".format(E, "°")) 
print("----------------")

#Hallemos r

r = a*(1 - e*(np.cos(E*np.pi/180)))

print("El radio vector esta dado por {0} {1}".format(r, "u.a."))


print("----------------")

#Hallemos anomalia verdadera (theta)
#(esta en radianes)
theta = 2*np.arctan(np.sqrt((1+e)/(1-e)) * np.tan(np.deg2rad(E/2))) 

print("La anomalía verdadera esta dada por {0} {1}".format(np.rad2deg(theta)+360, "°"))

print("----------------")

#Con esto podemos hallar x, y, z:

x = r*(np.cos(BigOmega)*np.cos(SmallOmega+theta) - np.sin(BigOmega)*np.sin(SmallOmega+theta) * np.cos(i))
y = r*(np.sin(BigOmega)*np.cos(SmallOmega+theta) + np.cos(BigOmega)*np.sin(SmallOmega+theta) * np.cos(i))
z = r*(np.sin(SmallOmega + theta) * np.sin(i))

print("x esta dada por {0} {1}".format(x, "ua"))
print("y esta dada por {0} {1}".format(y, "ua"))
print("z esta dada por {0} {1}".format(z, "ua"))

#Control:

rControl = np.sqrt(math.pow(x,2)+math.pow(y,2)+math.pow(z,2))

print("rControl esta dada por {0} {1}".format(rControl, "ua"))

print("----------------------------------------------------------")

#Ahora pasemos x,y,z de eclipticas a equatoriales
#dado epsilon
eps = 23.43927944
eps = np.deg2rad(eps) #epsilon en radianes

#Coordenadas equatoriales helicentricas de la tierra
x_q = x 
y_q = y*np.cos(eps) - z*np.sin(eps)
z_q = y*np.sin(eps) + z*np.cos(eps)


#Para pasar a coordenadas geocentricas equatoriales del sol simplemente multiplicamos por -1

x_q = -x_q
y_q = -y_q
z_q = -z_q

#Ahora con esto pasamos de rectangulares a esfericas

alph = np.arctan(y_q/x_q)
rh = np.sqrt(math.pow(x_q,2) + math.pow(y_q,2) + math.pow(z_q,2))
dt = np.rad2deg(np.arcsin(z_q/rh))

alph = atansi(np.rad2deg(alph),y_q,x_q)

#grados a horas minutos y segundos
print(alph)
print(dt)
h = np.floor(alph/15)
m = np.floor(((alph/15)-h)*60)
s = (((alph/15)-h)*60 -m)*60
alpha=[h,m,s]

#grados a grados minutos y segundos

dt = - dt
ht = np.floor(dt)
mt = np.floor((dt-ht)*60)
st = ((dt-ht)*60 - mt)*60
delta = [ht,mt,st]


print("la ascencion recta esta dada por {0} {1} {2} {3} {4} {5}".format(alpha[0], "h",alpha[1], "m",alpha[2], "s"))
print("la distancia geocentrica esta dada por {0} {1}".format(rh, "ua"))
print("la declinación esta dada por {0} {1} {2} {3} {4} {5}".format(-delta[0], "°",delta[1], "m",delta[2], "s"))
```

```python
import pandas as pd
import numpy as np
import math 

#Ejercicio 5
# Un objeto de masa despreciable que gira alrededor del Sol tiene una velocidad
#en el perihelio de 60 km/s y una velocidad en el afelio de 30 km/s. Determinar
#su excentricidad
#Valores que tenemos:
vel_p = 60 #(En km/s)
vel_a = 30 #(En km/s)
K = 0.01720209908
 
#Con 
# 1 u.a. = 149597870.691 km
# 1 día = 86400 segundos
#pasemos a u.a/d
 
vel_p= (vel_p /149597870.691) * 86400
 
vel_a= (vel_a /149597870.691) * 86400

#ahora utilizando que vel_p = K(sqrt((1+e)/a(1-e))) y que vel_a = K(sqrt((1-e)/a(1+e)))
#podemos hallar a y e multiplicando y dividiendo vel_p y vel_a, luego:

a = math.pow(K,2)/(vel_p * vel_a)

e = (vel_p - vel_a)/(vel_p + vel_a)

print("la excentricidad esta dada por {0} {1}".format(e, ""))

print("el semieje mayor esta dado por {0} {1}".format(a, "au"))
```
