```python
#en caso de que se pase de 360 o sea negativo
def gradnorm(alpha):
  if alpha > 360:
    alpha = alpha - 360
  elif alpha < 0:
    alpha = alpha + 360
  return alpha

#Verificar el tangente inverso 
def atansi(alpha,a,b):
  if a*b < 0 and b<0:
    alpha = alpha + 180
  elif a*b < 0 and b>0:
    alpha = alpha + 360
  elif a+b < 0:
    alpha = alpha + 180
  return alpha  

```
