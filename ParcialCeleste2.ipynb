```python
#Para verificar la tangente
def arctansafe(alpha,a,b):
  if a*b < 0 and b<0:
    alpha = alpha + 180
  elif a*b < 0 and b>0:
    alpha = alpha + 360
  elif a+b < 0:
    alpha = alpha + 180
  return alpha
```

```python
import pandas as pd
import numpy as np
import math

#Ejercicio 1
#hallé mas de lo que debía 
K = 0.01720209908

#Datos dados, Coordenadas (vector posición) eclipticas
x=-1.03244991
y=2.36352624
z=0.26487849

#Vector velocidad

xdot=-0.00968498
ydot=-0.00498515
zdot=0.00162665

#Magnitud de vectores
r = np.sqrt(math.pow(x,2) + math.pow(y,2) + math.pow(z,2)) 
print("r: {0} {1}".format(r, ""))

v = np.sqrt(xdot**2 + ydot**2 + zdot**2)
print("v: {0} {1}".format(v, ""))

#Miramos si es cónica o esta en trayectoria rectilinea

h1= (y*zdot) - (z*ydot)
h2= (z*xdot) - (x*zdot)
h3= (x*ydot) - (y*xdot)

h = np.sqrt(h1**2 + h2**2 + h3**2)

print("h: {0} {1}".format(h, ""))

#Excentricidad (K**2 = miu?)

e1 = (-1/K**2)*((h2*zdot) - (h3*ydot)) - (x/r)
e2 = (-1/K**2)*((h3*xdot) - (h1*zdot)) - (y/r)
e3 = (-1/K**2)*((h1*ydot) - (h2*xdot)) - (z/r)

e = np.sqrt(e1**2 + e2**2 + e3**2)
tes = e1*h1+e2*h2+e3*h3
print("Comprobante test: {0} {1}".format(tes, ""))

print("e: {0} {1}".format(e, ""))

#Miremos que tipo de órbita tiene 

if(e==1):
    print("parabola")
elif(e<1):
    print("elipse")
elif(e>1):
    print("Hipérbola")
Q = (r*v**2)/(K**2)
#Hallamos rdot (rpunto) para mirar en que cuadrante es:

rcompdot = (x*xdot + y*ydot + z*zdot)/r
print("r punto es: {0} {1}".format(rcompdot, ""))
#HELP: rdot < 0 -> se acerca al sol   (TERCER O CUARTO CUADRANTE) SE CUMPLE ESTE CASO
    #  rdot > 0 -> se aleja del sol   (PRIMER O SEGUNDO CUADRANTE)
#Calculemos a

print("Q: {0} {1}".format(Q, ""))

a = r/(2-Q)

print("a: {0} {1}".format(a, ""))

#Calculamos i

i = np.arccos(h3/h)

i = np.rad2deg(i)+180

print("i: {0} {1}".format(i, ""))

#Calculamos omegaGrande

BigOm = np.arctan(h1/-h2)

BigOm = np.rad2deg(BigOm)

BigOm = arctansafe(BigOm,h1,-h2)

print("OMEGA: {0} {1}".format(BigOm, ""))

#Calculamos omega pequeño

SmallOmega = np.arctan((e3*h)/(-e1*h2 + e2*h1))

SmallOmega = np.rad2deg(SmallOmega)

SmallOmega = arctansafe(SmallOmega,(e3*h),(-e1*h2 + e2*h1))
print("smallOmega: {0} {1}".format(SmallOmega, ""))

#Calculemos t, para ello:
#Primero caclulemos theta

theta = np.arccos((1/e)*((a*(1-e**2)/r) - 1))

#rdot < 0, luego tercer o cuarto cuadrate
theta = np.rad2deg(theta)+180

print("theta: {0} {1}".format(theta, ""))

```
